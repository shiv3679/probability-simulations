```python
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib.animation as animation
```

We have to simulate a recusrion relation given by the following formula   
$X_{t+1} = a_tX_{t} + b_t$ 
where $a_t$ is given by $a_t = \left(1 +  \displaystyle{\frac{a}{T_t}} \right)$ and $b_t = \left(\displaystyle{\frac{b}{T_t}} \right)$ and $T_t = at + b(t+1)$ with $t$ as the time parameter. 

```python
# getting the initial inputs for a and b = 1
# a = int(input("Enter a: "))
# b = int(input("Enter b: "))

t = np.linspace(0,100,100)    # number of iterations the simulation will run for

def T_t(a,b,t = []):
    # number of iterations the simulation will run for
    t = np.linspace(0, 100, 100)
    T_t = np.empty(len(t))
    for i in range(len(t)-1):
        T_t[i] = a*t[i] + b*(t[i+1])
    return T_t


```

```python
print(t)   # printing the time values
```

```python
T = T_t(1,1,t)   # calling the function to get the T_t values
print(T)
```
